<h3>Chapter 2: Merge files</h3>
<p>This chapter reads and merges all web scraping files into one.</p>

```python
# Define file name to merge
files_to_merge = [
    'cities_climate/SCOR_Cities_Climate_CDO.csv',
    'cities_climate/SCOR_Cities_Climate_WWO.csv'
]
```

```python
# Setup empty dataframe to join the datasets
result = {
            'State': [],
            'City': [],
            'Month': [],
            'Avg. Temperature °C (°F)': [],
            'Min. Temperature °C (°F)': [],
            'Max. Temperature °C (°F)': [],
            'Precipitation / Rainfall mm (in)': [],
            'Humidity(%)': [],
            'Rainy days (d)': [],
            'avg. Sun hours (hours)': []
            }

df_result = pd.DataFrame(result)
```

```python
# Join datasets and remove duplicates (if any)
for file in files_to_merge:
    df_new = pd.read_csv(file)
    df_result = pd.concat([df_result, df_new])
    
df_result.drop_duplicates()
```

```python
# Export csv as checkpoint for further processing
df_result.to_csv('cities_climate/SCOR_Cities_Climate.csv')
```
