<h3>Chapter 2: Cluster assignment to submission file</h3>
<p>This chapter reads the clustered files and submission template, then prepares a file for submission.</p>

```python
# List of file directories
clustered_files = [
    "datasets_clustered/kharif_clustered_done.csv",
    "datasets_clustered/rabi_clustered_done.csv",
    "datasets_clustered/state_kharif_clustered_done.csv",
    "datasets_clustered/state_rabi_clustered_done.csv"
]

submission_template_files = [
    "./../../03_Submission/03_Prediction/GP_Pred_Kharif_ID.csv",
    "./../../03_Submission/03_Prediction/GP_Pred_Rabi_ID.csv",
    "./../../03_Submission/03_Prediction/States_Pred.csv"
]

submission_files = [
    "./../../03_Submission/03_Prediction/GP_Pred_Kharif_GR21_ID.csv",
    "./../../03_Submission/03_Prediction/GP_Pred_Rabi_GR21_ID.csv",
    "./../../03_Submission/03_Prediction/States_Pred_GR21.csv"
]
```

<hr/>
<b><u>For Kharif - Free Level</u></b>

```python
# Select clustering parameters
cluster_column = 'kmn_clusters'
```

```python
# Store clustered datasets in a dataframe
df = {}
df_clusters = pd.read_csv(clustered_files[0], delimiter=";",index_col = 0)
df_template = pd.read_csv(submission_template_files[0])                                                  
```

```python
# Split IDs on clusters file
df_id = df_clusters['ID'].str.split("_", expand=True)
df_clusters['ID'] = df_id[0] + "_" + df_id[2]
df_clusters
```

```python
# Add ID on template
df_template['ID'] = df_template['State'] + "_" + df_template['District']
df_template
```

```python
# Join df_template and df_cluster, then fil in Cluster column
df_result = pd.merge(df_template, df_clusters[['ID', cluster_column]], how='left', on='ID')
df_result['Cluster'] = df_result[cluster_column]
df_result['Cluster'] = df_result['Cluster'].fillna(max(df_result[cluster_column])+1)
df_result = df_result.drop(columns=['ID', cluster_column])
```

```python
# Export result
df_result.to_csv(submission_files[0])
```

<b><u>For Kharif and Rabi - State Level</u></b>

```python
# Select clustering parameters
cluster_column = 'msc_clusters'
```

```python
# Store clustered datasets in a dataframe
df = {}
df_clusters_kharif = pd.read_csv(clustered_files[2], delimiter=";",index_col = 0)
df_clusters_rabi = pd.read_csv(clustered_files[3], delimiter=";",index_col = 0)
df_template = pd.read_csv(submission_template_files[2])
```

```python
df_clusters_rabi
```

```python
# Join df_template and df_cluster for kharif and rabi, then fil in Cluster column
# For Kharif
df_result = pd.merge(df_template, df_clusters_kharif[['ID', cluster_column]], how='left', left_on='State', right_on='ID')
df_result['Kharif'] = df_result[cluster_column]
df_result['Kharif'] = df_result['Kharif'].fillna(max(df_result[cluster_column])+1)
df_result = df_result.drop(columns=['ID', cluster_column])

# For Rabi
df_result = pd.merge(df_result, df_clusters_rabi[['ID', cluster_column]], how='left', left_on='State', right_on='ID')
df_result['Rabi'] = df_result[cluster_column]
df_result['Rabi'] = df_result['Rabi'].fillna(max(df_result[cluster_column])+1)
df_result = df_result.drop(columns=['ID', cluster_column])

df_result
```

```python
# Export result
df_result.to_csv(submission_files[2])
```
